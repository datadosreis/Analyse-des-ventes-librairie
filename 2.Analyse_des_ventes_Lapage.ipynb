### CA Annuel

```python
CA_annuel = ventes_final.groupby(pd.Grouper(key='date', freq='A'))['price'].sum()
CA_annuel
#CA par année
```

Seule l'année 2022 est complète. Les données ne sont donc pas comparables au niveau annuel.
- l'année 2021 débute au 01/03/2021
- l'année 2023 s'arrête au 28/02/2023

### CA Mensuel

```python
CA_mensuel = ventes_final.groupby(pd.Grouper(key='date', freq='M'))['price'].sum().reset_index()
CA_mensuel.rename(columns={'price': 'CA mensuel'}, inplace=True)
CA_mensuel["CA mensuel"] = round(CA_mensuel["CA mensuel"] / 1000,2)
CA_mensuel

#CA mensuel sur l'ensemble de la période (24 mois) converti en K€
```

```python
CA_mensuel.plot(x='date',y='CA mensuel',marker='o',figsize=(16,5))
plt.title('Evolution mensuelle du CA',fontsize=14)
plt.xlabel("Période", fontsize=12)
plt.ylabel("CA en K€", fontsize=12)
plt.grid(axis='y')
plt.ylim(300,550)
plt.show()
```

L'analyse de l'évolution mensuelle du CA indique clairement une anomalie sur le CA d'octobre 2021.
Il va nous falloir investiguer davantage sur cette baisse soudaine en regardant tout d'abord si l'une des 3 catégories est impactée ou s'il s'agit d'un problème plus global.

A noter également, une baisse importante sur février 2023.

### Evolution du CA par catégorie

```python
ventes_par_categ = ventes_final.pivot_table(index=pd.Grouper(key='date', freq='M'), columns='categ', values='price', aggfunc='sum')
ventes_par_categ.columns = ["0", "1", "2"]

ventes_par_categ[["0", "1", "2"]].plot(figsize=(16,5))
plt.title('Evolution mensuelle du CA par catégorie',fontsize=14)
plt.xlabel("Période", fontsize=12)
plt.ylabel("CA en €", fontsize=12)
plt.grid(axis='y')
plt.show()
```

L'anomalie semble se situer sur les ventes de la catégorie 1.
Nous allons donc analyser spécifiquement les données du mois d'octobre 2021.
Par ailleurs, la baisse sur février 2023 concerne toutes les catégories et ne semble donc pas être une anomalie

```python
ventes_oct21 = ventes_final.pivot_table(index=pd.Grouper(key='date', freq='D'), columns='categ', values='price', aggfunc='sum').reset_index()
ventes_oct21.columns= ["date","0", "1", "2"]
ventes_oct21 = ventes_oct21[ventes_oct21["date"].isin(pd.date_range('2021-10-1', '2021-10-31'))]
ventes_oct21['1'] = ventes_oct21['1'].fillna(0)

ventes_oct21.plot(x='date',figsize=(16,5))
plt.title("Evolution du CA d'octobre 21 par catégorie",fontsize=14)
plt.xlabel("Période", fontsize=12)
plt.ylabel("CA en €", fontsize=12)
plt.grid(axis='y')
plt.show()
```

Il semble y avoir eu un problème dans la remontée des ventes des produits de catégorie 1 sur la période du 02 octobre 2021 au 27 octobre 2021 inclus. En effet, aucune vente n'a été comptabilisée sur cette période.

En conséquence, il faudra exclure les données du mois d'octobre 2021 afin de ne pas perturber les différentes analyses qui suivront.

### Analyse des prix

```python
listing_prix = ventes_final[['id_prod', 'price','categ']]
listing_prix = listing_prix.groupby('id_prod').mean()
listing_prix['categ'] = listing_prix['categ'].astype("int")
```

```python
cat_0 = listing_prix.loc[listing_prix['categ'] == 0]
cat_1 = listing_prix.loc[listing_prix['categ'] == 1]
cat_2 = listing_prix.loc[listing_prix['categ'] == 2]
```

#### Global

```python
pricing_analysis(listing_prix['price'])
```

Au global, les prix varient de 0.62€ à 300€ toutes catégories confondues.
Le prix moyen est de 21.76€ et le prix médian est de 13.04€.
La distribution n'est clairement pas symétrique avec un skewness empirique de 3.81, les prix tendent donc vers la droite (c'est à dire vers des prix élevés).

#### Catégorie 0

```python
pricing_analysis(cat_0['price'])
```

Sue la catégorie 0, les prix varient de 0.62€ à 40.99€.
Le prix moyen est de 11.71€ et le prix médian est de 10.30€.
La distribution reste étalée vers la droite (skewness = 0.85) mais la répartition est plus symétrique que lorsque l'on regarde sur toutes les catégoties confondues.

Nous observons quelques outliers mais nous ne les considérons pas comme des valeurs aberrantes.

#### Catégorie 1

```python
pricing_analysis(cat_1['price'])
```

Sur la catégorie 1, les prix varient de 2€ à 80.99€.
Le prix moyen est de 25.50€ et le prix médian est de 22.99€.
La distribution reste étalée vers la droite (skewness = 0.82) mais la répartition est plus symétrique que lorsque l'on regarde sur toutes les catégoties confondues.

Nous observons 10 outliers mais nous ne les considérons pas comme des valeurs aberrantes.

#### Catégorie 2

```python
pricing_analysis(cat_2['price'])
```

Sur la catégorie 2, les prix varient de 30.99€ à 300€.
Le prix moyen est de 107.64€ et le prix médian est de 101.25€.
La distribution reste étalée vers la droite (skewness = 0.95) mais la répartition est plus symétrique que lorsque l'on regarde sur toutes les catégoties confondues.

Nous observons 6 outliers mais encore une fois, nous ne les considérons pas comme des valeurs aberrantes.

**Conclusion**

L'analyse des prix montre des niveaux de prix très différents selon les catégories:

- Prix moyen catégorie 0 = 11.71€
- Prix moyen catégorie 1 = 25.50€
- Prix moyen catégorie 2 = 107.64€

Nous pouvons donc en conclure que les produits des 3 catégories sont très différents les uns des autres.
Pour rappel, la gamme de produits la plus large concerne la catégorie 0 avec plus de 2200 produits. A l'inverse, la gamme de la catégorie 2 n'est composé "que" de 236 produits.

#### Pareto des ventes

```python
pareto = top_ventes_qty.sort_values(ascending=False).to_frame() #tri par total_sales (décroissant)
pareto["% des ventes"] = round((pareto / pareto.sum()) * 100,2) #création de la colonne % des ventes
pareto["% Cumulé"] = pareto["% des ventes"].cumsum() # cumsum calcule la somme cumulée
pareto.rename(columns={'session_id': 'ventes'}, inplace=True)
pareto.head(10)
```

```python
print("80% des ventes sont réalisées avec",pareto.loc[pareto['% Cumulé'] < 80 ].shape[0],"produits.")
```

## Lien entre le genre d’un client et les catégories des livres achetés

**Hypothèses et niveau de confiance**

*Quelques hypothèses:*

- Il y a un lien entre le genre d'un client et les catégories de livres achetés
- Les hommes achètent davantage de livres des catégories 0 et 2
- Les femmmes achètent davantage de livres de la catégorie 1
- Il n'y a pas de lien entre le genre et les catégories

Nous procéderons au test des hypothèses suivantes:

- **Hypothèse nulle H0** = Le genre d'un client et les catégories des livres achetés sont deux variables indépendantes *(si p-value > 0.05)*
- **Hypothèse alternative H1** = Le genre d'un client et les catégories des livres achetés ne sont pas deux variables indépendantes *(si p-value < 0.05)*

Nous choisissons un niveau de confiance de 95 %, c'est-à-dire que nous rejetterons l'hypothèse nulle en faveur de l'alternative si la p-value est inférieure à 0,05.

Il s'agit ici d'étudier la relation entre la variable "genre du client" et "catégorie des livres achetés" qui sont toutes deux des variables qualitatives. Nous effectuerons donc un test d'indépendance du Chi2.

**Représentation des données**

Nous sommes ici dans le cas de l'analyse des corrélations entre deux variables qualitatives, nous optons donc pour une première représentation des données sous forme de tableau de contingence.

```python
contingency_table = pd.crosstab(ventes["sex"],ventes["categ"])
print("Tableau de contingence:\n",contingency_table)
```

```python
#Représentation graphique du tableau de contingence
fig=plt.figure()
contingency_table.T.plot(kind='barh', ylabel='frequence')

plt.title("Genre du client vs catégorie des livres achetés")
```

**Test statistique: le test d'indépendance du Chi2**

```python
#Test d'indépendance du Chi2
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi2 =",chi2)
print("P-value =",p)
print("Degree of Freedom =",dof)
print("Valeurs attendues:\n",expected)

if p<0.05:
    print("Rejet de l'hypothèse nulle H0")
else:
    print("L'hypothèse nulle H0 est acceptée")
```

**Interprétation des résultats**

<div class="alert alert-block alert-info">
    La p-value étant largement inférieure au seuil de significativité de 5%, nous <b>rejetons donc l'hypothèse nulle</b> et concluons donc qu'il y a bien un lien entre le genre d'un client et les catégories des livres achetés.
</div>

## Lien entre l’âge des clients et le montant total des achats

### Test de normalité

Les tests statistiques qui suivent vont tous concerner la variable 'âge'. Afin de savoir si nous utiliserons des tests paramétriques ou non-paramétriques, nous décidons de procéder à un **test de normalité** de cette variable pour voir si les observations suivent une loi normale.

**Hypothèses et niveau de confiance**

Les hypothèses sont les suivantes:

- **Hypothèse nulle H0** = les données suivent une loi normale

- **Hypothèse alternative H1** = les données ne suivent pas une loi normale

Nous choisissons un niveau de confiance de 95 %, c'est-à-dire que nous rejetterons l'hypothèse nulle en faveur de l'alternative si la p-value est inférieure à 0,05.

Nous commençons donc par représenter graphiquement la variable âge sous forme d'histogramme, puis nous utiliserons le test d'adéquation de Kolmogorov-Smirnov.

**Représentation des données**

```python
#Représentation graphique
plt.figure(figsize=(16,5))
sns.histplot(data=ventes['age'],bins=40, kde=True)
plt.title("Répartition des ventes par âge", fontsize=12)
plt.show()
```

**Test statistique: le test de Kolmogorov-Smirnov**

```python
#Test de Kolmogorov-Smirnov
statistic, p = stats.kstest(ventes['age'],stats.norm.cdf)
print("Kolmogorov-Smirnov =",statistic)
print("P-value =",p)

if p<0.05:
    print("Rejet de l'hypothèse nulle H0")
else:
    print("L'hypothèse nulle H0 est acceptée")
```

**Interprétation des résultats**

<div class="alert alert-block alert-info">
    
La p-value est inférieure à notre seuil de significativité de 5%, nous <b>rejetons donc l'hypothèse nulle</b> en faveur de l'alternative: les données ne sont pas distribuées selon la loi normale. 

Nous utiliserons donc des <b>tests non-paramétriques</b>.
    
</div>

### Corrélation de Spearman

**Hypothèses et niveau de confiance**

*Quelques hypothèses:*

- Il y a un lien entre l'âge d'un client et le montant total des achats
- Les clients les plus âgés sont les clients qui dépensent le moins
- Les clients entre 30 et 50 ans sont les clients qui dépensent le plus
- Il n'y a pas de lien entre l'âge et le montant total des achats

Nous procéderons au test des hypothèses suivantes:

- **Hypothèse nulle H0** = pas de corrélation entre l'âge et le montant total des achats

- **Hypothèse alternative H1** = corrélation entre l'âge et le montant total des achats

Nous choisissons un niveau de confiance de 95 %, c'est-à-dire que nous rejetterons l'hypothèse nulle en faveur de l'alternative si la p-value est inférieure à 0,05.

Nous sommes ici dans la situation où nous souhaitons analyser la relation entre 2 variables quantitatives.
La distribution de la variable "âge" ne suivant pas une loi normale, nous allons donc opter pour un test non paramétrique: le test de **corrélation de Spearman.** Avant cela, nous allons effectuer une représentation graphique des données via un diagramme de dispersion.

**Représentation des données**

```python
achats_par_age = ventes.groupby("age")['price'].sum().reset_index()

#Représentation graphique
plt.figure(figsize=(16,5))
plt.plot(achats_par_age["age"],achats_par_age["price"],'o')
plt.xlabel("Âge")
plt.ylabel("Montant total des achats \n (en €)")
plt.title("Montant total des achats par âge")
plt.show()
```

**Test statistique: le test de corrélation de Spearman**

```python
#Corrélation de Spearman
correlation, p = stats.spearmanr(achats_par_age["age"],achats_par_age["price"])
print("Spearman =",correlation)
print("P-value =",p)

if p<0.05:
    print("Rejet de l'hypothèse nulle H0")
else:
    print("L'hypothèse nulle H0 est acceptée")
```

**Interprétation des résultats**

<div class="alert alert-block alert-info">
    
La p-value est inférieure à notre seuil de significativité de 5%, nous <b>rejetons donc l'hypothèse nulle</b> en faveur de l'alternative: il y a bien une corrélation négative entre l'âge et le montant total des achats.

## Lien entre l’âge des clients et les catégories des livres achetés

**Hypothèses et niveau de confiance**

*Quelques hypothèses:*

- Il y a un lien entre l'âge d'un client et les catégories des livres achetés
- Les produits de la catégorie 2 intéressent davantage les clients jeunes
- La catégorie 1 est celle qui rassemble l'ensemble des âges
- Les personnes âgées ne s'intéressent pas aux produits de catégories 0 et 2
- Il n'y a pas de lien entre l'âge d'un client et les catégories des livres achetés

Nous procéderons au test des hypothèses suivantes:

- **Hypothèse nulle H0** = pas de lien entre l'âge et les catégories des livres achetés.
- **Hypothèse alternative H1** = lien entre l'âge et les catégories des livres achetés.

Nous choisissons un niveau de confiance de 95 %, c'est-à-dire que nous rejetterons l'hypothèse nulle en faveur de l'alternative si la p-value est inférieure à 0,05.

Nous sommes ici dans la situation où nous souhaitons analyser la corrélation entre 1 variable quantitative (âge) et 1 variable qualitative à 3 modalités (catégories 0,1,2). La distribution de la variable "âge" ne suivant pas une loi normale, nous optons donc pour un test non paramétrique qui sera le test de **Kruskall-Wallis**.
Nous effectuons une représentation graphique des données via un boxplot.

**Représentation des données**

```python
meanprops = {'marker':'x', 'markeredgecolor':'red','markerfacecolor':'firebrick'}

sns.boxplot(data=ventes, x="categ" , y="age",fliersize=False,showmeans=True, meanprops=meanprops)
plt.title("Répartion des âges par catégorie")
plt.show()
```

**Test statistique: le test de Kruskall-Wallis**

```python
categ_0 = ventes.loc[ventes['categ'] == 0]
categ_1= ventes.loc[ventes['categ'] == 1]
categ_2= ventes.loc[ventes['categ'] == 2]

#Test de Kruskall-Wallis
statistic, p = stats.kruskal(categ_0['age'],categ_1['age'],categ_2['age'])
print("Kruskall-Wallis =",statistic)
print("P-value =",p)

if p<0.05:
    print("Rejet de l'hypothèse nulle H0")
else:
    print("L'hypothèse nulle H0 est acceptée")
```

# Conclusion

Voici les points clés que nous pouvons mettre en lumière suite à l'analyse des ventes de la librairie Lapage:

**Indicateurs clés:**

- La société a réalisé un CA de 11,8 millions d'euros sur la période du 01/03/2021 au 28/02/2023 inclus
- Une grande partie des données de ventes du mois d'octobre 2021 pour les produits de la catégorie 1 sont manquantes
- Les niveaux de prix sont très différents selon les catégories de produits
- 61% des ventes (en volume) sont réalisées avec des produits de la catégorie 0
- La répartition des ventes est égale entre les hommes et les femmes (50,50)
- L'âge moyen des clients se situent aux alentours de 44 ans (avec des clients allant de 18 à 93 ans)
- Les produits de catégorie 2 sont uniquement achetés par des clients entre 18 et 30 ans
- 4 clients réalisent des chiffres d'affaires supérieurs à 100K€, probablement des clients professionnels
- Le panier moyen par session est de 2 articles pour un montant moyen de 34.63 €

**Tests statistiques:**

- Le test d'indépendance du chi2 démontre clairement un lien entre le genre et les catégories des livres achetés.


- Il existe une corrélation entre l'âge du client et montant total des achats.
    - Il semblerait qu'à partir de 40 ans, le montant total des achats diminue avec l'âge


- Il existe une corrélation entre l'âge du client et la fréquence d'achat.
    - Encore une fois, il semble qu'après 40 ans la fréquence d'achat diminue avec l'âge


- Il existe également une corrélation entre l'âge du client et la taille du panier moyen.
    - Les clients âgés entre 30 et 50 ans sont ceux qui ont les paniers les plus importants


- Le test de Kruskall-Wallis nous a démontré qu'il existe clairement un lien entre l'âge et les catégories des livres achetés.
    - Les clients entre 18 et 30 ans sont les seuls à acheter les produits de catégorie 2
    - Les produits de la catégorie 1 rassemblent l'ensemble des âges
    - Les personnes de plus de 70 ans ne semblent pas enclins à acheter des produits de la catégorie 0


Ces données nous permettent d'avoir une première idée quant au profil des clients de la librairie en ligne Lapage.
